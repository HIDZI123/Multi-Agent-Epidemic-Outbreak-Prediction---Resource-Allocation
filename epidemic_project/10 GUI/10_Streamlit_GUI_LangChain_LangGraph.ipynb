# Notebook 10: Streamlit GUI for Notebook 08 and 09\n
\n
This notebook starts a Streamlit dashboard that explains and visualizes the working of:\n
- Notebook 08 (LangChain multi-agent flow)\n
- Notebook 09 (LangGraph workflow)\n
\n
The original notebooks are not modified.

```python
import os
import sys
import subprocess
from pathlib import Path

base_dir = Path.cwd()
gui_dir = base_dir / '10 GUI'
if not gui_dir.exists():
    gui_dir = Path.cwd()

req_path = gui_dir / 'requirements.txt'
print('Installing dependencies from:', req_path)
subprocess.check_call([sys.executable, '-m', 'pip', 'install', '-q', '-r', str(req_path)])
print('Dependencies installed.')
```

```python
import subprocess
from pathlib import Path

gui_dir = Path.cwd() / '10 GUI'
if not gui_dir.exists():
    gui_dir = Path.cwd()

app_path = gui_dir / 'streamlit_app.py'
print('Starting Streamlit app:', app_path)
process = subprocess.Popen(['streamlit', 'run', str(app_path)])
print('Streamlit started. Open the URL shown in the cell output/terminal.')
print('To stop it, interrupt kernel or terminate the streamlit process.')
```

## Notes\n
\n
- Run this notebook from inside the `10 GUI` folder for the simplest path resolution.\n
- If your environment cannot find `streamlit`, restart kernel after dependency install.\n
- The dashboard supports LangChain mode, LangGraph mode, and side-by-side comparison mode.
